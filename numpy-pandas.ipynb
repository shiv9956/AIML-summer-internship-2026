```
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, %, **, //): ")
num2 = float(input("Enter second number: "))
if operator == "+":
    print(f"Result: {num1} + {num2} = {num1 + num2}")

elif operator == "-":
    print(f"Result: {num1} - {num2} = {num1 - num2}")

elif operator == "*":
    print(f"Result: {num1} * {num2} = {num1 * num2}")

elif operator == "/":
    if num2 == 0:
        print("Error: Division by zero is not allowed.")
    else:
        print(f"Result: {num1} / {num2} = {num1 / num2}")

elif operator == "%":
    if num2 == 0:
        print("Error: Cannot find remainder when dividing by zero.")
    else:
        print(f"Result: {num1} % {num2} = {num1 % num2}")

elif operator == "**":
    print(f"Result: {num1} ** {num2} = {num1 ** num2}")

elif operator == "//":
    if num2 == 0:
        print("Error: Floor division by zero is not allowed.")
    else:
        print(f"Result: {num1} // {num2} = {num1 // num2}")

else:
    print("Invalid operator! Please use +, -, *, /, %, **, or //.")
```

```
balance = float(input("Enter starting balance: $"))
deduction = float(input("Enter amount to deduct: $"))
if deduction <= 0:
    print("Error: Deduction amount must be greater than 0.")

elif deduction > balance:
    print(f"Declined: Insufficient funds. (Balance: ${balance:.2f}, Attempted: ${deduction:.2f})")

else:
    balance -= deduction
    print(f"Success: ${deduction:.2f} deducted. New balance: ${balance:.2f}")
```

```
num = int(input("enter a number: "))
num = abs(num)
if 0 <= num <= 9:
    print(f"{num} is single digit")
elif 10 <= num <= 99:
    print(f"{num} is double digit")
elif 100 <= num <= 999:
    print(f"{num} is triple digit")
else:
    print(f"{num} is 4 or more digits (or out of bounds)")
```

```
#most optimal
import math

num = int(input("enter a number: "))
num = abs(num)

# Special case: log10(0) is mathematically undefined, so we handle 0 manually
if num == 0:
    digit_count = 1
else:
    # Example: math.log10(150) is ~2.176 -> math.floor(2.176) is 2 -> 2 + 1 = 3 digits
    digit_count = math.floor(math.log10(num)) + 1

# Now we can dynamically label the digit type
if digit_count == 1:
    print(f"{num} is a single digit")
elif digit_count == 2:
    print(f"{num} is a double digit")
elif digit_count == 3:
    print(f"{num} is a triple digit")
else:
    print(f"{num} has {digit_count} digits")
```
