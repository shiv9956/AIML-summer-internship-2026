```python
tb(modi["Tweet"][10]).sentiment
"""*   **Polarity**: This value indicates the emotional charge of the text,
ranging from -1 (very negative) to +1 (very positive).
A polarity of 0.0 suggests a neutral sentiment.

*   **Subjectivity**: This value indicates how subjective or objective the text is,
 ranging from 0.0 (very objective) to 1.0 (very subjective). Objective text expresses facts,
 while subjective text expresses opinions, feelings, or beliefs.
"""
```

```python
tb(rahul["Tweet"][2]).sentiment
```

```python
modi["Tweet"] = modi["Tweet"].astype(str)
rahul["Tweet"] = rahul["Tweet"].astype(str)

def find_polarity(review):
  return tb(review).sentiment.polarity
```

```python
modi["Polarity"] = modi["Tweet"].apply(find_polarity)
rahul["Polarity"] = rahul["Tweet"].apply(find_polarity)
```

```python
modi
```

```python
rahul
```

```python
modi["Label"] = np.where(modi["Polarity"] > 0,"Positive","Negative")
modi["Label"][modi["Polarity"] == 0] = "Neutral"

rahul["Label"] = np.where(rahul["Polarity"] > 0,"Positive","Negative")
rahul["Label"][rahul["Polarity"] == 0] = "Neutral"
```
