```python
import gradio as gr
from google import genai

# 1. Initialize modern GenAI client
API_KEY = "your-gemini-api-key"  # <--- Replace with your key
client = genai.Client(api_key=API_KEY)

# 2. Create a persistent Gemini chat session
chat_session = client.chats.create(model="gemini-2.5-flash")

# 3. Simple, state-managed prediction function
def predict(message, history):
    # Pass ONLY the new message to Gemini's active chat session
    response = chat_session.send_message(message)
    return response.text

# 4. Launch UI
gr.ChatInterface(predict).launch(share=True)
```
