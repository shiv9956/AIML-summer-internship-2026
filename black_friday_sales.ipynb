```python
data = pd.DataFrame({'Ratio':[len(df[df['Gender'] == 'M']) , len(df[df['Gender'] == 'F'])]},
                    index = ['Male',"Female"]) # Create a DataFrame to store the count of males and females
data.plot.pie(y = 'Ratio' , figsize = (6,6), autopct = '%.1f') # Plot a pie chart showing the distribution of Gender, with percentages formatted to one decimal place
```

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize = (6,6)) # Plot a bar chart showing the count of purchases by Gender
```

```python
df['Gender'].value_counts() # Get the counts of unique values in the 'Gender' column
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (6,6),title ='Purchase distribution by age') # Plot a bar chart showing the number of purchases for each Age group
```

```python
data = df.groupby('Age')['Product_ID'].nunique().reset_index() # Group by Age and count the number of unique products purchased
data.columns = ['Age','Products'] # Rename columns for clarity
```

```python
data.plot.bar(x = 'Age',figsize = (8,6)) # Plot a bar chart showing the number of unique products purchased by Age group
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12,6),
                              title= 'Purchase distribution by Age') # Plot another bar chart for purchase distribution by Age, showing the count of records per age group
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (12,6),
                              title= 'Amount spend by Age') # Plot a bar chart showing the total purchase amount for each Age group
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'bar', figsize = (12,6),
                              title= 'Average Purchase Amount by Age') # Plot a bar chart showing the average purchase amount for each Age group
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'pie', autopct ='%0.1f', figsize = (12,6),
                              title= 'Average Purchase Amount by Age') # Plot a pie chart showing the average purchase amount by Age, with percentages
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct ='%0.1f') # Plot a pie chart showing the distribution of Marital Status
```
