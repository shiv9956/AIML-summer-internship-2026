```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\Users\shivd\Downloads\zomato.csv") # Read the CSV file into a pandas DataFrame
df.head(2) # Display the first 2 rows of the DataFrame
```

```python
df.drop(["url","address","phone","reviews_list"],axis=1,inplace=True) # Drop specified columns from the DataFrame
df.head(2) # Display the first 2 rows of the modified DataFrame
```

```python
df.drop(["url","address","phone","reviews_list"],axis=1,inplace=True)
df.head(2)
```

```python
df.columns # Display the column names of the DataFrame
```

```python
df.drop_duplicates(inplace=True) # Remove duplicate rows from the DataFrame
df.shape # Display the shape (number of rows and columns) of the DataFrame
```

```python
df["rate"].unique() # Display unique values in the 'rate' column
```

```python
#removing "NEW","-" and "/5" from rate column
def handlerate(value) :
    if (value=="NEW" or value=="-"):
        return np.nan # Replace 'NEW' or '-' with NaN (Not a Number)
    else:
        value =str(value).split('/')
        value = value[0]
        return float(value)

df["rate"]= df["rate"].apply(handlerate) # Apply the handlerate function to the 'rate' column
df["rate"].head() # Display the first 5 processed values of the 'rate' column
```

```python
df["rate"].unique() # Display unique values in the 'rate' column after cleaning
```

```python
df.fillna({"rate": df["rate"].mean()}, inplace=True) # Fill NaN values in 'rate' column with its mean
df['rate'].isnull().sum() # Check for any remaining null values in the 'rate' column
```

```python
df.info() # Display a concise summary of the DataFrame
```

```python
df.dropna(inplace=True) # Drop rows with any NaN values from the DataFrame
df.head(2) # Display the first 2 rows of the DataFrame after dropping NaN values
```

```python
df.shape # Display the shape (number of rows and columns) of the DataFrame
```

```python
df.rename(columns={"approx_cost(for two people)":"Cost2plates","listed_in(type)":"Type"},inplace=True) # Rename columns for better readability
df.head(2) # Display the first 2 rows of the DataFrame with renamed columns
```

```python
df["location"].unique() # Display unique values in the 'location' column
```

```python
df["Cost2plates"].unique() # Display unique values in the 'Cost2plates' column
```

```python
#remove comma from cost2plates
def handlecomma(value):
    value=str(value)
    if ',' in value :
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates']= df['Cost2plates'].apply(handlecomma) # Apply the handlecomma function to clean 'Cost2plates' column
df['Cost2plates'].unique() # Display unique values in the 'Cost2plates' column after cleaning
```

```python
df.head() # Display the first 5 rows of the DataFrame
```

```python
rest_types = df['rest_type'].value_counts(ascending=False) # Calculate value counts for 'rest_type' column
rest_types
```

```python
rest_type_lessthan1000 = rest_types[rest_types < 1000] # Filter rest types with counts less than 1000
rest_type_lessthan1000
```

```python
# if data less than 1000 in rest type, sum them all and count as others
def handle_rest_type(value):
    if value in rest_type_lessthan1000:
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type) # Apply the function to group less frequent rest types

df['rest_type'].value_counts() # Display value counts of 'rest_type' after handling
```

```python
#cleaning cuisines
cuisines= df['cuisines'].value_counts(ascending=False) # Calculate value counts for 'cuisines' column

cuisines_lessthan100= cuisines[cuisines < 100] # Filter cuisines with counts less than 100

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines) # Apply the function to group less frequent cuisines

df['cuisines'].value_counts() # Display value counts of 'cuisines' after handling
```
